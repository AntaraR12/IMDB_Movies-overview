lets gather more information on directors 

```python
df1 = df.groupby(["director_name"])[["year", "title"]].aggregate({"year": ["min", "max"], "title": "count"})
```

lets convert multi-index to one

```python
df1.columns= ["_".join (i) for i in df1.columns]
```

```python
df1
```

lets change the index 

```python
df1.reset_index()
```

lets find the career span of directors 

```python
df1["career_span"] = df1["year_max"] - df1["year_min"]
df1
```

lets find the productivity of the directors 

```python
df1["productivity"] = df1["title_count"]/df1["career_span"] 
df1
```

```python
df1.sort_values(by= ["productivity"], ascending= False).head(5)
```
